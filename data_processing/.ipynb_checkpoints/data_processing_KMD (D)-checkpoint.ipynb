# Data processing KMD data
Due to the fact that only part of the KMD weather data was provided and only near the end of the project, currently no KMD data is being used for the model. However, if more data was provided this notebook would be able to visualize it so if the dataset is complete in the end in can be combined with the CHIRPS remote sensing dataset.

Known issue: Only importing single date, useful for visualisation. Rpelace name with * to make the .glob functions import all files 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import glob
import platform
import pathlib  
import sys
import netCDF4 as nc
import geopandas as gpd
import rasterio
from rasterio.mask import mask
from shapely.geometry import Point
from pathlib import Path
```

```python
# Load in the precipitation data that is located in the GIT folder itself
path_data_folder = Path(r"../data/3_precip")
path_precip = path_data_folder.joinpath(r"chirps_2018_2022_28.45_43.55_-5.05_5.45.nc")

print(f"Precipitation data will be loaded from the {path_precip} folder")
```

```python
# If errors occur here please refer to the readme file or to the file_imports.py folders. 

cwd = pathlib.Path().resolve()
src = cwd.parent
data = src.parent.parent.parent
root = src.parent
OS_type = platform.system()
sys.path.append(str(src))
sys.path.append(str(root))
from utils.file_imports import *

shape_path = file_paths(root, catchments = True)[0]
```

```python
#Load in the dataset contained in the .NetCDF file(s)
data_files = glob.glob(str(path_precip))
datasets = {}

# Loop through the Netcdf file and assign the correct coordinates to be able to plot the dataset as Geopandas \
# dataframe. Prints the xarray file in the end for manual inspection of the file structure.

for file_path in data_files:
    
    # Extract the file identifier from the file name to be added as the time coordinate (the file name)
    # originally contained the date only as part of the file name. 
    file_name = os.path.basename(file_path)  
    file_identifier = file_name.split('_')[0] 
    dataset = xr.open_dataset(file_path)
    dataset = dataset.assign_coords(file_identifier=file_identifier)
    
    # Add the dataset to the dictionary with the file identifier as the key
    datasets[file_identifier] = dataset

print(dataset)
```

```python
#Loading in the shapefile of the Tanabasin so it can be used to visualize the availability of the KMD data 

shape_file = os.path.join(shape_path, '*.gpkg')
shape_files = glob.glob(shape_file)
gdfs = []

for file in shape_files:
    gdf = gpd.read_file(file)
    gdfs.append(gdf)
```

```python
#Plotting the available precipitation data on the shapefile of the Tana Basin. 
# Time specifies the data on which we want the data to be plotted.
combined_gdf = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))
fig, ax = plt.subplots(figsize=(10, 8)) 
combined_gdf.plot(ax=ax, color='bisque', edgecolor='black')
time = '2018-04-11'
precip_data = datasets['chirps']['precip'].sel(time='2018-04-11')
precip_data.plot(ax=ax, x='longitude', y='latitude', cmap='viridis', alpha=0.8)

# Plot the resulting precipitation dataset
ax.set_xlim(36.5, 42)
ax.set_ylim(-3.5, 1)
plt.title(f'CHIRPS rainfall data on {time}')
plt.show()
```
