```python
from superflexpy.implementation.elements.hbv import PowerReservoir              #import of reservoir
from superflexpy.implementation.elements.gr4j import UnitHydrograph1            #import lag function
from superflexpy.implementation.root_finders.pegasus import PegasusPython       #numerical solver   
from superflexpy.implementation.numerical_approximators.implicit_euler import ImplicitEulerPython   #numerical solver
from superflexpy.framework.unit import Unit             #model elements
from superflexpy.framework.node import Node             #model elements
from superflexpy.framework.network import Network       #model elements
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

FIRST EXAMPLE

```python
solver_python = PegasusPython()
approximator = ImplicitEulerPython(root_finder=solver_python)
```

```python
reservoir = PowerReservoir(
    parameters={'k': 0.01, 'alpha': 2.0},
    states={'S0': 10.0},
    approximation=approximator,
    id='R'
)
```

```python
precipitation = np.array([6.5, 3, 0, 0, 2, 2, 4, 8, 2, 0, 0, 2, 1, 0, 4, 0, 1, 0, 4, 6, 0, 0,  ])
reservoir.set_timestep(1.0)
reservoir.set_input([precipitation])
```

```python
output = reservoir.get_output()[0]
reservoir_state = reservoir.state_array[:, 0]
```

```python
fig, ax = plt.subplots(2, 1, sharex=True, figsize=(10, 6))
ax[0].bar(x=range(len(precipitation)), height=precipitation, color='blue')
ax[1].plot(range(len(precipitation)), output, color='blue', lw=2, label='Outflow')
ax_bis = ax[1].twinx()
ax_bis.plot(range(len(precipitation)), reservoir_state, color='red', lw=2, ls='--', label='Reservoir state')
```

SECOND EXAMPLE

```python
#The elements

reservoir_2 = PowerReservoir(
    parameters={'k': 0.01, 'alpha': 2.0},
    states={'S0': 10.0},
    approximation=approximator,
    id='R2'
)

lag_function = UnitHydrograph1(
    parameters={'lag-time': 2.3},
    states={'lag': None},
    id='lag-fun'
)
```

```python
#To combine the different elements 

unit_1 = Unit(
    layers=[[reservoir], [lag_function]],
    id='unit-1'
)
```

```python
unit_1.set_timestep(1.0)
unit_1.set_input([precipitation])
x = 5
```

```python
# Lag function is added to the output of the unit
output_2 = unit_1.get_output()
```

```python
r_state = unit_1.get_internal(id='R', attribute='state_array')[:, 0]
r_output = unit_1.call_internal(id='R', method='get_output', solve=False)#[0]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10,9))

# plot first figure
ax1.bar(x=range(len(precipitation)), height=precipitation, color='blue', label='precipitation')
ax1.legend()
ax1.set_xlabel('Time [days]')
ax1.set_ylabel("precipitation [mm/day]")

# plot second figure
ax_bis = ax2.twinx()
ax2.plot(range(len(precipitation)), r_output[0], color='blue', lw=2, label='Outflow')
ax2.set_ylabel('Outflow')  # Set y-label for the second subplot
ax_bis.set_ylabel('Reservoir State')
ax_bis.plot(range(len(precipitation)), r_state, color='red', lw=2, ls='--', label='Reservoir state')

# legend second figure
handles, labels = [], []
for ax in [ax2, ax_bis]:
    for h, l in zip(*ax.get_legend_handles_labels()):
        handles.append(h)
        labels.append(l)

ax2.legend(handles, labels, loc='upper right')

# plot third figure
ax3.plot(range(len(precipitation)), output_2[0], color = 'blue', lw=2, label='Time lag')
ax3.legend()
ax3.set_xlabel("Time [days]")
ax3.set_ylabel("Outflow [mm/day]")
```

```python
unit_1.reset_states()
```

EXAMPLE 3

```python
reservoir_3 = PowerReservoir(
    parameters={'k': 0.01, 'alpha': 2.0},
    states={'S0': 10.0},
    approximation=approximator,
    id='R'
)

lag_function_2 = UnitHydrograph1(
    parameters={'lag-time': 2.3},
    states={'lag': None},
    id='lag-fun-2'
)

unit_2 = Unit(
    layers=[[reservoir_3], [lag_function_2]],
    id='unit-2'
)

unit_3 = Unit(
    layers=[[reservoir_3]],
    id='unit-3'
)
```

```python
node_1 = Node(
    units=[unit_2, unit_3],
    weights=[0.7, 0.3],
    area=10.0,
    id='node-1'
)
```

```python
node_1.set_timestep(1.0)
node_1.set_input([precipitation])
```

```python
output_3 = node_1.get_output()[0]
```

```python
output_unit_2 = node_1.call_internal(id='unit-2', method='get_output', solve=False)[0]
output_unit_3 = node_1.call_internal(id='unit-3', method='get_output', solve=False)[0]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10,9))

# plot first figure
ax1.bar(x=range(len(precipitation)), height=precipitation, color='blue', label='precipitation')
ax1.legend()
ax1.set_xlabel('Time [days]')
ax1.set_ylabel("precipitation [mm/day]")

# plot second figure
ax2.plot(range(len(precipitation)), output_unit_2, color='blue', lw=2, label="Unit 1")
ax2.plot(range(len(precipitation)), output_unit_3, color='red', lw=2, ls='--', label="Unit 2")
ax2.set_ylabel('Outflow [mm/day]')
ax2.legend()

# plot third figure
ax3.plot(range(len(precipitation)), output_3, color = 'blue', lw=2, label='Time lag')
ax3.legend()
ax3.set_xlabel("Time [days]")
ax3.set_ylabel("Outflow [mm/day]")
```

EXAMPLE 4

```python
reservoir_4= PowerReservoir(
    parameters={'k': 0.01, 'alpha': 2.0},
    states={'S0': 10.0},
    approximation=approximator,
    id='R4'
)

lag_function_4 = UnitHydrograph1(
    parameters={'lag-time': 2.3},
    states={'lag': None},
    id='lag-fun'
)

unit_4_1 = Unit(
    layers=[[reservoir_4], [lag_function_4]],
    id='unit-4-1'
)

unit_4_2 = Unit(
    layers=[[reservoir_4]],
    id='unit-4-2'
)

node_4_1 = Node(
    units=[unit_4_1, unit_4_2],
    weights=[0.7, 0.3],
    area=10.0,
    id='node-4-1'
)

node_4_2 = Node(
    units=[unit_4_1, unit_4_2],
    weights=[0.3, 0.7],
    area=5.0,
    id='node-4-2'
)

node_4_3 = Node(
    units=[unit_4_2],
    weights=[1.0],
    area=3.0,
    id='node-4-3'
)
```

```python
net = Network(
    nodes=[node_4_1, node_4_2, node_4_3],
    topology={
        'node-4-1': 'node-4-3',
        'node-4-2': 'node-4-3',
        'node-4-3': None
    }
)
```

```python
node_4_1.set_input([precipitation])
node_4_2.set_input([precipitation * 0.5])
node_4_3.set_input([precipitation + 1.0])
```

```python
net.set_timestep(1.0)
```

```python
output_4 = net.get_output()
```

```python
output_unit_4_1_node_4_1 = net.call_internal(id='node-4-1_unit-4-1',
                                         method='get_output',
                                         solve=False)[0]

output_unit_4_1_node_4_2 = net.call_internal(id='node-4-1_unit-4-2',
                                         method='get_output',
                                         solve=False)[0]

output_unit_4_1_node_4_3 = net.call_internal(id='node-4-3',
                                         method='get_output',
                                         solve=False)[0]
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,6))

# plot first figure
ax1.bar(x=range(len(precipitation)), height=precipitation, color='blue', label='precipitation')
ax1.legend()
ax1.set_xlabel('Time [days]')
ax1.set_ylabel("precipitation [mm/day]")

ax2.plot(range(len(precipitation)), output_unit_4_1_node_4_1, color='blue', lw=2, label='Node 1')
ax2.plot(range(len(precipitation)), output_unit_4_1_node_4_2, color='red', lw=2, label='Node 2')
ax2.plot(range(len(precipitation)), output_unit_4_1_node_4_3, color='orange', lw=2, label='Node3')
ax2.legend()
ax2.set_ylabel('Outflow [mm/day]')
```
