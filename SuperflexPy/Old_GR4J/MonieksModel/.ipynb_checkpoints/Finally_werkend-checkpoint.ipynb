```python
path = os.getcwd()
data_folder = os.path.join("source_data", "text_files")
asc_folder = os.path.join(path, 'asc_files')

DEM=np.genfromtxt(os.path.join(asc_folder, 'demmuthk.asc'),  dtype=float, autostrip=True)
slope=np.genfromtxt(os.path.join(asc_folder,'slopemuthk.asc'), dtype=float, autostrip = True)
hand= np.genfromtxt(os.path.join(asc_folder, 'handmuthk.asc'), dtype = float, autostrip = True)
basin= np.genfromtxt(os.path.join(asc_folder, 'basin.asc'), dtype = float, autostrip = True)
LULC = np.genfromtxt(os.path.join(asc_folder, 'lulc_muthk.asc'), dtype = float, autostrip = True)

#plot DEM
plt.figure(1)
plt.title('DEM')
DEM[DEM==-9999]=np.nan
plt.imshow(DEM, cmap='hsv')
plt.colorbar()
plt.show()
 
#plot HAND
plt.figure(2)
plt.figure('HAND')
hand[hand==-9999]=np.nan
plt.imshow(hand, cmap='hsv')
plt.colorbar()
plt.show()

#make landscape classification
hillslope = np.array(slope) > 4
plateau = (np.array(hand) > 1) & (np.array(slope) <= 4)
wetland = (np.array(hand) <= 1) & (np.array(slope) <= 4)
basin = np.array(basin)>= 0

hillslope_per = float(np.sum(hillslope))/float(np.sum(basin))
wetland_per = float(np.sum(wetland))/float(np.sum(basin))
plateau_per = float(np.sum(plateau))/float(np.sum(basin))

landscapes=np.zeros((482,573))
landscapes[plateau]=1
landscapes[hillslope]=2
landscapes[wetland]=3

#plot landscapes
cmap = mpl.colors.ListedColormap(['white', 'red', 'green', 'blue'])
bounds=[0,1,2,3,4]
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

plt.figure(3)
plt.title('Landscapes')
plt.imshow(landscapes, cmap=cmap, norm=norm)
plt.colorbar()
plt.show()

#plot LULC
plt.figure(4)
plt.title('LULC')
LULC[LULC==-9999]=np.nan
plt.imshow(LULC, cmap='hsv')
plt.show()

print(f'Percentage hillslope = {hillslope_per:.4f}')
print(f'Percentage wetland = {wetland_per:.4f}')
print(f'Percentage plateau = {plateau_per:.4f}')
```

```python
forcing=np.genfromtxt('asc_files/IMD_1975_2013.txt',  dtype=float, autostrip=True)

years = int(len(forcing[:,0]) / 365.25)
fert_start = np.zeros((len(forcing), 1))
for i in range(years):
    value = 1
    fert_start[60+(365*i)] = value
    fert_start[152+(365*i)] = value
    
df = pd.DataFrame(fert_start)
Firr  = np.linspace(120, 240, 10)
factor = 1.02

for i in range(10):
    df[i] = pd.DataFrame(fert_start*Firr[i])
    value = 1
    factor = 1.00005
    for j in range(len(forcing)):
        value = value * factor
        df[i][j] = df[i][j] * value
    
df.columns = ['W1', 'W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8', 'W9', 'W10']
df.to_csv('Fert_scen.csv', index = False)
```

```python
forcing = np.column_stack((forcing[:,:], np.array(df.iloc[:,9])))
frac1 = 0.5       # part that will be irrigated

k = 0.01 # decay of fertilizer

                  #             Imax  Ce      Sumax beta Pmax    Kf    k
ParPlateau   =        np.array([2.06, 0.9902, 200,  1.5, 0.0849, 0.45, k])   
ParHillslope_crop =   np.array([2.06, 0.9902, 100,  1.5, 0.8,    0.45, k])
ParHillslope_forest = np.array([2.06, 0.9902, 300,  1.5, 0.8,    0.45, k])
ParWetland   =        np.array([2.06, 0.9902, 77,   1.5, 0.65,   0.45, k])

              # Ks Tlag
ParCatchment = np.array([0.0281, 2.21,])

landscape_per=np.array([plateau_per, hillslope_per, wetland_per])

Qm, Ea, Ctot, Def = FLEXtopo(ParPlateau, ParHillslope_crop, ParHillslope_forest, ParWetland, ParCatchment, forcing, landscape_per, frac1)

plt.plot(Ctot)

Qo = forcing[:,0]
Prec = forcing[:,1]
Ep = forcing[:,2]

fig, ax1 = plt.subplots(figsize=(12,6))

ax11 = ax1.twinx()
ax11.invert_yaxis()
ax12 = ax1.twinx()
ax11.spines.right.set_position(('axes', 1.05))

ax1.plot(range(2600, 3700), Qo[2600:3700], color = "C1", alpha = 0.6, label = 'Qobserved')
ax1.plot(range(2600, 3700), Qm[2600:3700], color = 'C2', alpha = 0.6, label = 'Qmodeled')
ax11.bar(range(2600,3700), Prec[2600:3700], color = "C0", alpha = 0.6, label = "Precipitation")
ax12.plot(range(2600, 3700), Ctot[2600:3700], ls = '--', color= "C3", alpha = 0.6, label = 'Ctot')

ax11.set_ylim(150,0)
ax1.set_ylim(-2,150)
ax12.set_ylim(-0.1, 10)

ax1.set_ylabel('Q [mm/day]')
ax11.set_ylabel('P [mm/day]')
ax12.set_ylabel('Concentration [?]')
ax1.set_xlabel('Time [days]')

ax1.legend(loc = 'center right')
ax11.legend(loc = 'upper right')
ax12.legend(loc = 'lower right');
```
